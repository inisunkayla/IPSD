# PIECHART

Piechart dibawah ini menunjukkan persentase rating yang diberikan customer terhadap barang yang dibeli. Diketahui 29.9% pelanggan memberikan rating 3 bintang.

```python
most_wasted = data2.groupby(["Rating"])[['Customer ID']].sum()

plt.figure(figsize=(5, 5))
plt.pie(most_wasted['Customer ID'], labels=most_wasted.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  
plt.title('Persentase Rating Customer September 2023-September 2024')
plt.show()
```

# DIAGRAM BATANG

Diagram batang dibawah menunjukkan banyaknya barang yang terjual berdasarkan tipenya pada September 2023 - September 2024. Smartphone menjadi tipe barang yang paling banyak dibeli pada toko tersebut.

```python
summ = data2.groupby(["Product Type"])[['Quantity']].sum()

plt.figure(figsize=(8, 6))
bars = plt.bar(summ.index, summ['Quantity'], color='skyblue')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')

plt.xlabel('Tipe Produk')
plt.ylabel('Banyak Produk')
plt.title('Total Tipe Produk Terjual')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
